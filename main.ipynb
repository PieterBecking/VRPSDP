```python
# Import files distance_matrix_1.csv and distance_matrix_2.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Import the data with delimeter as ; in stad of comma
df1 = pd.read_csv('distance_matrix_1.csv', delimiter=';', index_col=0)
df2 = pd.read_csv('distance_matrix_2.csv', delimiter=';', index_col=0)


# Print the first 5 rows of the dataframe
print(df1.head())
print(df2.head())

# After loading the data
df1.columns = df1.index
df2.columns = df2.index


# the data represents the distance between different nodes. The nodes are represented by the index of the dataframe in the first column and the first row. The distance between the nodes is represented by the value in the dataframe. The distance is symmetric, so the distance from node i to node j is the same as the distance from node j to node i.
# Now we will visualize the data using a heatmap. We will use the seaborn library to create the heatmap.

# Create a heatmap of the data
sns.heatmap(df1, annot=True)
plt.show()

sns.heatmap(df2, annot=True)
plt.show()

# now visualize in a node and edge graph
import networkx as nx
G1 = nx.from_pandas_adjacency(df1)
G2 = nx.from_pandas_adjacency(df2)

nx.draw(G1, with_labels=True)
plt.show()
nx.draw(G2, with_labels=True)
plt.show()


```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Assuming G1 and G2 are your graphs

# Draw G1 with edge labels
pos = nx.spring_layout(G1)  # positions for all nodes
nx.draw(G1, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G1, 'weight')
nx.draw_networkx_edge_labels(G1, pos, edge_labels=edge_labels)
plt.show()

# Draw G2 with edge labels
pos = nx.spring_layout(G2)  # positions for all nodes
nx.draw(G2, pos, with_labels=True)
edge_labels = nx.get_edge_attributes(G2, 'weight')
nx.draw_networkx_edge_labels(G2, pos, edge_labels=edge_labels)
plt.show()
```
