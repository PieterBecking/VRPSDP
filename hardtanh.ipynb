```python
# create a plot of a hard tanh function
import numpy as np
import matplotlib.pyplot as plt

# define the hard tanh function
def hard_tanh(x):
    if x < -1:
        return -1
    elif x > 1:
        return 1
    else:
        return x
    
# create a range of x values
x = np.linspace(-2, 2, 100)
# calculate the y values
y = [hard_tanh(i) for i in x]

# create the plot
plt.plot(x, y)
plt.title('Hard Tanh Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
