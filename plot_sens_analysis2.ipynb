```python
import matplotlib.pyplot as plt

# Data
data = {
    "Vehicle Range Change": [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0],
    "Objective Function Value": [3252.588, 2920.2200000000003, 2920.2200000000003, 2772.1585000000005, 2714.328, 2714.328, 2626.209, 2626.209, 2626.209, 2624.473, 2624.473],
    "Runtime": [648.9427709579468, 160.96724104881287, 1015.3451449871063, 1565.985985994339, 836.5765399932861, 1082.672513961792, 230.41044998168945, 294.3192410469055, 375.65367007255554, 289.50208806991577, 326.92772912979126]
}

fig, ax1 = plt.subplots()

# Plotting Objective Function Value
ax1.set_xlabel('Vehicle Range Change')
ax1.set_ylabel('Objective Function Value', color='blue')
ax1.plot(data["Vehicle Range Change"], data["Objective Function Value"], 'bo-', label='Objective Function Value')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a second y-axis for the Runtime completely grey
ax2 = ax1.twinx()
ax2.set_ylabel('Runtime', color='grey')
ax2.plot(data["Vehicle Range Change"], data["Runtime"], 'o-', color='lightgrey', label='Runtime')
ax2.tick_params(axis='y', labelcolor='grey')

# Adding a title
plt.title('Vehicle Range Change vs Objective Function Value and Runtime')

# Combining legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.12), shadow=True, ncol=2)

# Showing the plot
plt.show()
```
